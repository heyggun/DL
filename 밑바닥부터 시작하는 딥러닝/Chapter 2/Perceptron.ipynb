# Perceptron

    신경망(딥러닝)의 기원이 되는 알고리즘
    다수의 신호를 입력으로 받아 하나의 신호를 출력함
    복수의 입력 신호에 각각 고유한 가중치를 부여하며, 가중치는 신호가 겨로가에 주는 영향열을 조절하는 요소로 작용
    가중치가 클수록 해당 신호가 그만큼 더 중요함을 뜻함

### single-layer perceptron

    AND gate ( 두 개의 입력값이 1 이여야 출력값이 1, 나머지는 0)

```python
def AND_gate(x1, x2):
    w1,w2,theta= 0.5,0.5,0.7
    tmp = x1*w1 + x2*w2
    
    if tmp <= theta:
        return 0
    else:
        return 1
```

```python
print(AND_gate(0,0))
print(AND_gate(0,1))
print(AND_gate(1,0))
print(AND_gate(1,1))
```

**가중치와 편향**

```python
import numpy as np

x = np.array([0,1]) # input
w = np.array([0.5,0.5]) # weight
b = -0.7 #bias

print(w*x)
print(np.sum(w*x))
print(np.sum(w*x)+b)
```

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(x*w)+b
    if tmp <=0:
        return 0
    else:
        return 1
    
    
```

    NAND gate ( 두 개의 입력값이 1 이면 출력값이 0, 나머지는 1)

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(x*w)+b
    if tmp <=0:
        return 0
    else:
        return 1
    
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5,0.5])
    b = -0.2
    tmp = np.sum(w*x) +b
    if tmp <=0:
        return 0
    else:
        return 1
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))

print('--------')

print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

    XOR Gate : 두 개의 입력이 같으면 0, 다르면 1
    


    하얀색 원(출력값 0), 검정색 원(출력값 1) 을 직선 하나로 나눈 것 불가능하고 두 개의 선이 필요함
    => 다층 퍼셉트론의 등장

```python
def XOR(x1, x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1, s2)
    
    return y
```

```python
print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```

    - 퍼셉트론은 입출력을 갖춘 알고리즘이다. 입력을 주면 정해진 규칙에 따른 값을 출력한다.
    - 퍼셉트론에서는 '가중치'와 '편향'을 매개변수로 설정한다.
    - 퍼헨트론으로 AND, OR 등의 논리 회로를 표현할 수 있다.
    - XOR 게이트는 단층 퍼셉트론으로는 표현할 수 없다.
    - 2층 퍼셉트론을 이용하면 XOR 게이트를 표현할 수 있다.
    - 단층 퍼셉트론은 직선형 영역만 표현할 수 있고, 다층 퍼셉트론은 비선형 영역도 표현할 수 있다.
    - 다층 퍼셉트론은 (이론상) 컴퓨터를 표현할 수 있다.
